```python
"""This script for Machine learning algorithm for video games sales."""

# Pandas library is used for reading data in table format.
# DecisionTreeClassifier for training the ML model.
# Train_test_split splits the data for two purposes(training & testing).
# Accuracy_score is used for verifiying ML model accuracy.
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
#from sklearn.preprocessing import LabelEncoder
```

```python
# Read and verify the dataset
video_game_sales_set = pd.read_csv('vgsales.csv')
# Outputs the whole dataset in table form.
video_game_sales_set
```

```python
# Cleaning the data

video_game_sales_set.dropna(inplace=True) # Removes rows with any missing values
video_game_sales_set.dropna(axis=1)  # Removes columns with any missing values
video_game_sales_set.drop_duplicates()
#label_encoder = LabelEncoder()
#video_game_sales_set['Name'] = label_encoder.fit_transform(video_game_sales_set['Name'])
# Verify data does not have duplicates
duplicates = video_game_sales_set.duplicated().sum()
empty_entries = video_game_sales_set.isnull().sum()
# if duplicates is 0 and all columns have zero nulls the data is clean
duplicates
empty_entries
```

```python
# Divide input and output from the dataset.
input_for_model = video_game_sales_set.drop(columns = ['Global_Sales']) # x
output_for_model = video_game_sales_set['Global_Sales'] # y

# Generate Test sets and training sets
# test_size is out of 1.0 . 0.2 means 20% of the set is for testing and 80% for training
xtraining_set, xtest_set, ytraining_set, ytest_set = train_test_split(input_for_model,output_for_model, test_size = 0.2)
```

```python
# Training the actual model

model = DecisionTreeClassifier()
model.fit(xtraining_set,ytraining_set)
```

```python
# test out the accuracy of the model
model_predictions = model.predict(xtest_set)
score = accuracy_score(ytest_set,model_predictions)
print(f"Model accuracy = {score*100} %")
```
